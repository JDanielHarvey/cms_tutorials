```python
pd.set_option('display.max_rows',200)
grp_country2 = df.groupby(['Country'])['EdLevel']
df2 = pd.DataFrame(grp_country2.apply(lambda x: x.str.contains('Master').sum()))
df2.head(200).sort_values(by='Country', ascending=True)
```

```python
tstt = pd.DataFrame(df['Country'].value_counts()).rename(columns={'Country':'Count'})
tstt.sort_index(ascending=True)
```

```python
grp_country2.apply(lambda x: x.str.contains('Master').sum())/df['Country'].value_counts()
```

```python
grp_ctry_master = grp_country2.apply(lambda x: x.str.contains('Master').sum())
grp_ctry_all = df['Country'].value_counts()
grp_concat = pd.concat([grp_ctry_master,grp_ctry_all], axis='columns', sort=False)

grp_concat.rename(columns={'EdLevel':'Masters','Country':'All'}, inplace=True)
```

```python
grp_concat['PctMaster'] = (grp_concat['Masters']/grp_concat['All'])
```
