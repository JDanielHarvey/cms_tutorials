```python
# -- filter mask
# -- filter mask can be dataframe but not if it's going to be applied back to selection
# -- notice the () parenthesis when setting the variable
# -- filter is a built-in keyword so use an abbreviation 

filt_age = (peeps['age'] > 35)
filt_name = (peeps['last_name'] == 'Winter')

filt_all = (peeps['age'] > 35) | (peeps['last_name'] == 'Winter')
```

```python
peeps[filt_age]
```

```python
peeps.loc[filt_age | filt_name]
```

```python
peeps.loc[filt_all]
```

```python
peeps.loc[~filt_age]
```
