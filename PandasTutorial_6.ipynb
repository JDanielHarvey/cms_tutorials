### Working with small data

```python
import pandas as pd
import os
```

```python
people = {
    "first": ["Joshua","Rebecca","Diane","John", "John", "James"],
    "middle": ["Daniel", "Lee", "Elizabeth", "Kohler", "Fake", "Fake"],
    "last": ["Harvey", "Winter", "McCabe", "Marsh", "Doe", "Doe"],
    "age": [31,27,60,58,10,10],
    "gender": ["m","female","female","m","m","m"]
}
```

```python
peeps = pd.DataFrame(people)
```

```python
peeps
```

```python
peeps.sort_values(by='age', ascending=False, inplace=True)
```

```python
peeps
```

```python
peeps.sort_values(by=['first','last'], ascending=True)
```

```python
# -- sorting by two columns with ascending specifications on both columns 
peeps.sort_values(by=['first','last'], ascending=[True, True])
```

```python
peeps.sort_index()
```

```python
# -- sort a single column
peeps['age'].sort_values()
```
